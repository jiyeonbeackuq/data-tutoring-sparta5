# 2-4) 결측값 처리 (삽입)

* **▶ orders - "배송시간" 결측값 처리를 다음과 같이 해보자**
* 1) 배송출발시간 ~ 배송시간의 평균 구하기 (mean_delivery)
* 2) 배송추정시간 = 배송출발시간 + mean_delivery
* 3) 배송시간 결측값을 배송추정시간으로 대체

* **▶ 1) 배송출발시간 ~ 배송시간의 평균 구하기 (mean_delivery)**
* **▶ orders - "배송출발시간", "배송시간"을 datetime으로 바꿔줘야함: pd.to_datetime**

```python
orders["배송출발시간"] = pd.to_datetime(orders["배송출발시간"])
orders["배송시간"] = pd.to_datetime(orders["배송시간"])
```

```python
mean_delivery = (orders['배송시간'] - orders['배송출발시간']).mean()
mean_delivery
```

* **▶ 2) 배송추정시간 = 배송출발시간 + mean_delivery**


```python
orders['배송추정시간'] = orders['배송출발시간'] + mean_delivery
```

```python
orders[orders["배송시간"].isna()]
```

* **▶ orders copy해두기 (orders2 생성)**

```python
orders2 = orders.copy()
```

* **▶ 3) 배송시간 결측값을 배송추정시간으로 대체**

```python
orders["배송시간"].fillna(orders["배송추정시간"], inplace = True)
```

# 2-5) 중복값 처리

**▶ orders - "주문ID"에서 앞 5글자만 추출해서 "ID" 만들기 - apply(lambda)**

```python
orders[:5]
```

```python
for index, i in enumerate(orders["주문ID"]):
  orders.loc[index, "ID"] = i[:5]
```

```python
orders['ID'] = orders['주문ID'].apply(lambda x:x[:5])
orders.head()
```

**▶ orders["ID"] 중복된 것 확인 (keep = False)**

```python
orders[orders['ID'].duplicated(keep = False)]
```

```python
orders.shape
```

**▶ orders["ID"] 중복값 지우기 (마지막 행 남기기)**

```python
orders.drop_duplicates(subset = "ID", keep = "last", inplace = True)
```

```python
orders[orders['ID'].duplicated(keep = False)]
```

**▶ orders - "고객ID"에서 앞 5글자만 추출해서 "CID" 만들기 - apply(lambda)**

```python
orders['CID'] = orders['고객ID'].apply(lambda x:x[:5])
```

```python
orders[orders['CID'].duplicated(keep = False)]
```
