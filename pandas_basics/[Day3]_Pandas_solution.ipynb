# 3-2) ÏãúÍ≥ÑÏó¥ Ï≤òÎ¶¨ (datetime)

**‚ñ∂ payments2 - "Ï£ºÎ¨∏ÏãúÍ∞Ñ" datetimeÌòïÌÉúÎ°ú type Î≥ÄÌôò**

```python
payments2['Ï£ºÎ¨∏ÏãúÍ∞Ñ'] = pd.to_datetime(payments2["Ï£ºÎ¨∏ÏãúÍ∞Ñ"])
```

```python
payments2['Ï£ºÎ¨∏ÏãúÍ∞Ñ']
```

**‚ñ∂payments2ÏóêÏÑú 'Ï£ºÎ¨∏ÏãúÍ∞Ñ' ÏπºÎüºÏùÑ Ïù¥Ïö©Ìï¥ Ïó∞ÎèÑÎßå Ï∂îÏ∂úÌïòÍ∏∞ ("Ï£ºÎ¨∏ÎÖÑÎèÑ" ÏÉùÏÑ±)**

```python
payments2['Ï£ºÎ¨∏ÎÖÑÎèÑ'] = payments2['Ï£ºÎ¨∏ÏãúÍ∞Ñ'].dt.year
```

**‚ñ∂payments2ÏóêÏÑú 'Ï£ºÎ¨∏ÏãúÍ∞Ñ' ÏπºÎüºÏùÑ Ïù¥Ïö©Ìï¥ ÏõîÎßå Ï∂îÏ∂úÌïòÍ∏∞ ("Ï£ºÎ¨∏Ïõî" ÏÉùÏÑ±)**

```python
payments2['Ï£ºÎ¨∏Ïõî'] = payments2['Ï£ºÎ¨∏ÏãúÍ∞Ñ'].dt.month
```

**‚ñ∂payments2ÏóêÏÑú 'Ï£ºÎ¨∏ÏãúÍ∞Ñ' ÏπºÎüºÏùÑ Ïù¥Ïö©Ìï¥ ÏùºÎßå Ï∂îÏ∂úÌïòÍ∏∞ ("Ï£ºÎ¨∏Ïùº" ÏÉùÏÑ±)**

```python
payments2['Ï£ºÎ¨∏Ïùº'] = payments2['Ï£ºÎ¨∏ÏãúÍ∞Ñ'].dt.day
```

```python
payments2
```

**‚ñ∂payments2ÏóêÏÑú "Ï£ºÎ¨∏Ïó∞Ïõî" Ïª¨ÎüºÏùÑ ÏÉùÏÑ±Ìï¥Î≥¥Ïûê**

```python
# Ï≤´Î≤àÏß∏ Î∞©Î≤ï (Î¨∏ÏûêÏó¥ Í≤∞Ìï©)
payments2['Ï£ºÎ¨∏ÎÖÑÎèÑ'].astype(str) + "-" + payments2['Ï£ºÎ¨∏Ïõî'].astype(str)
```

```python
# ÎëêÎ≤àÏß∏ Î∞©Î≤ï (dt.strftime Ïù¥Ïö©)
payments2["Ï£ºÎ¨∏Ïó∞Ïõî"] = payments2["Ï£ºÎ¨∏ÏãúÍ∞Ñ"].dt.strftime("%Y-%m")
```

# 3-3) Í∑∏Î£πÌôî

**‚ñ∂payments2ÏóêÏÑú "Í≤∞Ï†úÎ∞©Ïãù"Î≥Ñ "Í≤∞Ï†úÍ∏àÏï°" ÌèâÍ∑†ÏùÑ Íµ¨Ìï¥Î≥¥Ïûê (pay_mean) ÏÉùÏÑ±**

```python
pay_mean = payments2.groupby('Í≤∞Ï†úÎ∞©Ïãù')['Í≤∞Ï†úÍ∏àÏï°'].mean()
```

**‚ñ∂payments2ÏóêÏÑú "Í≤∞Ï†úÎ∞©Ïãù"Î≥Ñ "Í≤∞Ï†úÍ∏àÏï°" ÌèâÍ∑†, Ìï©Í≥Ñ Íµ¨Ìï¥Î≥¥Ïûê (pay_groupby) ÏÉùÏÑ±**

```python
pay_groupby = payments2.groupby('Í≤∞Ï†úÎ∞©Ïãù')['Í≤∞Ï†úÍ∏àÏï°'].agg(["mean","sum"])
```

**‚ñ∂Ïù∏Îç±Ïä§Î•º Ïª¨ÎüºÏúºÎ°ú Î∞îÍæ∏Í∏∞**

```python
pay_mean = pay_mean.reset_index()
```

```python
pay_mean
```

```python
pay_groupby.reset_index(inplace = True)
```

```python
pay_groupby
```

# 3-4) ÌîºÎ¥áÌåÖ

**‚ñ∂payments2ÏóêÏÑú Ï£ºÎ¨∏Ïó∞ÏõîÎ≥Ñ Ï£ºÎ¨∏ÏÉÅÌÉú Î∞è Í≤∞Ï†úÎ∞©ÏãùÏóê Îî∞Î•∏ ÌèâÍ∑† Í≤∞Ï†ú Í∏àÏï°, Í≥†Ïú† countÏùÑ ÏïåÏïÑÎ≥¥Ïûê. Ï¥ùÌï©Í≥Ñ Ï∂úÎ†• (result)**

```python
payments2
```

```python
result = payments2.pivot_table(values = "Í≤∞Ï†úÍ∏àÏï°",
                      index = ["Ï£ºÎ¨∏ÏÉÅÌÉú","Í≤∞Ï†úÎ∞©Ïãù"],
                      columns = ["Ï£ºÎ¨∏Ïó∞Ïõî"],
                      aggfunc = ["mean", "count"],
                      fill_value = 0,
                      margins = True)
```

```python
result.reset_index(inplace = True)
result
```
