# 문제2
아래 리스트(List)에서 apple, kiwi를 추출해서 result리스트로 출력하세요.
* 리스트 인덱싱을 사용하지 마세요.

x = ["grapes", "mango", "orange", "peach", "apple", "lime", "banana", "cherry", "tomato", "kiwi", "blueberry", "watermelon"]

출력 결과 : [apple, kiwi]

```python
x = ["grapes", "mango", "orange", "peach", "apple", "lime", "banana", "cherry", "tomato", "kiwi", "blueberry", "watermelon"]

result = []

# "" - False
# "abc" - True

for i in x:
  if i == "apple" or i == "kiwi":
    result.append(i)

for i in x:
  if i in ("apple", "kiwi"):
    result.append(i)
```

```python
import seaborn as sns

# Penguins 데이터셋 로드
penguins = sns.load_dataset("penguins")

# 'species'별 평균 bill_length_mm 계산 및 조건 필터링
penguins["mean_bill_length"] = penguins.groupby("species")["bill_length_mm"].transform("mean")
filtered_penguins = penguins[penguins["bill_length_mm"] < penguins["mean_bill_length"]]

# 새 열 추가: bill_length_mm과 bill_depth_mm의 가중합 계산
filtered_penguins["weighted_bill"] = filtered_penguins.apply(
    lambda row: row["bill_length_mm"] * 0.5 + row["bill_depth_mm"] * 0.3, axis=1
)

# weighted_bill이 25 이상인 행 필터링
final_result = filtered_penguins[filtered_penguins["weighted_bill"] >= 25]
final_result
```
