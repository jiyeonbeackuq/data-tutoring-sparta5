# 문제 1

```python
import seaborn as sns
import pandas as pd

# Iris 데이터셋 로드
iris = sns.load_dataset("iris")

# sepal_length의 평균 계산
mean_sepal_length = iris["sepal_length"].mean()

# 평균보다 큰 데이터와 작은 데이터를 각각 필터링
above_average = iris[iris["sepal_length"] > mean_sepal_length]
above_average["category"] = "above_average"

below_average = iris[iris["sepal_length"] < mean_sepal_length]
below_average["category"] = "below_average"

# 두 데이터프레임을 합치기
concat_df = pd.concat([above_average, below_average], ignore_index=True)

# 정답 출력
print(concat_df.iloc[50])
print(concat_df.iloc[120])
```

```python
import seaborn as sns
import pandas as pd

# Iris 데이터셋 로드
iris = sns.load_dataset("iris")

# species별 평균 petal_length와 petal_width 계산
species_avg = iris.groupby("species")[["petal_length", "petal_width"]].mean().reset_index()
species_avg.rename(columns={"petal_length": "avg_petal_length",
														"petal_width": "avg_petal_width"}, inplace=True)

# 원래 데이터셋과 병합
merged_df = pd.merge(iris, species_avg, on="species")

# 정답 출력
print(merged_df.iloc[10])
print(merged_df.iloc[50])
print(merged_df.iloc[120])
```

# 추가 문제

```python
# Tips 데이터셋 로드
import seaborn as sns
tips = sns.load_dataset("tips")

# 요일과 시간을 기준으로 그룹화 및 집계
grouped = tips.groupby(["day", "time"]).agg({"tip": "mean", "total_bill": "sum"})
display(grouped)

# 피벗 테이블 생성
pivot_table = grouped.reset_index().pivot_table(index="day",
																								columns="time",
																								values=["tip", "total_bill"])
pivot_table
```

# 데이터 변형 - 코드

```python
import pandas as pd
df = pd.DataFrame({"Name": ["Alice", "Bob"], "Age": [25, 30]}, index=["a", "b"])
df
```

```python

# 새 열 추가
df["City"] = ["Seoul", "Busan"]
df
```

```python

# 값 수정
df.loc["a", "Age"] = 26  # 'a'행, 'Age' 열 값 수정
df
```

```python

# 열 삭제
df = df.drop(columns=["City"])
df
```

```python

# 행 삭제
df = df.drop(index=["b"])
df
```

```python

# 데이터프레임 병합 (concat)
df2 = pd.DataFrame({"Name": ["Charlie"], "Age": [35]})
df = pd.concat([df, df2], axis = 0, ignore_index=True)

print(df)
print("-" * 30)
```

```python

# Age에 2를 곱하는 새 열 추가
df["Double Age"] = df["Age"].apply(lambda x: x * 2)
print(df)
print("-" * 30)
```

```python

# axis=0: 각 열(column)에 대해 함수를 적용
# 각 열의 합 계산
column_sums = df.apply(lambda col: col.sum(), axis=0)
print(column_sums)
print("-" * 30)
```

```python

# 각 행의 Name과 Age를 결합한 새 열 추가
# axis=1: 각 행(row)에 대해 함수를 적용.
df["Summary"] = df.apply(lambda row: f"{row['Name']} is {row['Age']} years old", axis=1)
print(df)
print("-" * 30)
```

```python

# 사용자 정의 함수 정의
def age_category(age):
    if age < 18:
        return "Underage"
    elif age < 65:
        return "Adult"
    else:
        return "Senior"

# Age 열에 사용자 정의 함수 적용
df["Age Category"] = df["Age"].apply(age_category)
print(df)
```

```python
# 데이터프레임 병합 (merge)
df3 = pd.DataFrame({"ID": [1, 2], "Name": ["Alice", "Bob"]})
df4 = pd.DataFrame({"ID": [1, 2], "Score": [90, 85]})
merged = pd.merge(df3, df4, on="ID")
print(merged)  # ID를 기준으로 병합
```

```python
grouped = df.groupby("City")["Age"].mean()
display(grouped)  # 'City'별 'Age' 평균값 계산

pivot_table = df.pivot_table(values="Age",
                             index="City",
                             aggfunc="mean")
pivot_table
```
